# Insertion Sort Explanation

## Example Array: [64, 34, 25, 12, 84]

### How Insertion Sort Works:

1. Start with the second element (34) and compare with elements to its left
2. Move larger elements to the right to make space for inserting the current element
3. Repeat this process for each element

### Step-by-Step Process:

**Initial Array:** [64, 34, 25, 12, 84]

**Pass 1:** Insert 34
- Compare 34 with 64
- 34 is smaller, shift 64 right
- Place 34 at correct position
- Result: [34, 64, 25, 12, 84]

**Pass 2:** Insert 25
- Compare 25 with previous elements
- Shift larger elements right
- Place 25 at correct position
- Result: [25, 34, 64, 12, 84]

**Pass 3:** Insert 12
- Compare 12 with previous elements
- Shift larger elements right
- Place 12 at correct position
- Result: [12, 25, 34, 64, 84]

**Pass 4:** Insert 84
- Compare 84 with previous elements
- 84 is already in correct position
- Result: [12, 25, 34, 64, 84]
"""


```python
def insertion_sort_with_counter(arr):
    comparisons = 0
    false = 0
    true = 0
    n = len(arr)
    
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        
        print(f"i={i}, key={key}")
        while j >= 0 and arr[j] > key:
            print(f"  j={j}, arr[j]={arr[j]} > {key}: True")
            comparisons += 1
            true += 1
            arr[j + 1] = arr[j]
            j -= 1
            
        if j >= 0:
            print(f"  j={j}, arr[j]={arr[j]} > {key}: False")
            false += 1
        comparisons += 1
        arr[j + 1] = key
    
    return arr, comparisons, true, false
```

```python
# Test the function
arr = [64, 34, 25, 12, 84]
sorted_arr, count, true, false = insertion_sort_with_counter(arr)
print(f"Sorted array: {sorted_arr}")
print(f"Number of comparisons: {count}")
print(f"True: {true}")
print(f"False: {false}")
```

```python
def show_insertion_sort_steps(arr):
    print("Insertion Sort Demonstration")
    print("----------------------------")
    print(f"Initial array: {arr}")
    print()
    
    n = len(arr)
    working_arr = arr.copy()
    
    for i in range(1, n):
        key = working_arr[i]
        j = i - 1
        
        print(f"Step {i}:")
        print(f"Current element (key): {key}")
        print(f"Current array: {working_arr}")
        
        # Show comparison process
        while j >= 0 and working_arr[j] > key:
            working_arr[j + 1] = working_arr[j]
            j -= 1
            print(f"Shifting {working_arr[j+1]} to the right")
            print(f"Array during shift: {working_arr}")
        
        working_arr[j + 1] = key
        print(f"Inserting {key} at position {j+1}")
        print(f"Array after insertion: {working_arr}")
        print("-" * 50)
        print()
    
    print("Final sorted array:", working_arr)
```

```python
# Demonstrate with the same array that will be used later
example_arr = [64, 34, 25, 12, 84]
show_insertion_sort_steps(example_arr)
```
